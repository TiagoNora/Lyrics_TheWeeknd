```python
import pandas as pd
```

```python
df = pd.read_csv("./tabelaTheWeeknd.csv")
df
```

```python
media_por_album = df.groupby('album')['valence'].mean()
media_por_album.sort_values(ascending=False)
```

```python
media_energy_por_album = df.groupby('album')['energy'].mean()
media_energy_por_album.sort_values(ascending=False)
```

```python
media_dance_por_album = df.groupby('album')['danceability'].mean()
media_dance_por_album.sort_values(ascending=False)
```

```python
df.loc[df.groupby('album')['valence'].idxmax()]\
                                .sort_values(by='valence', ascending=False)[['album', 'nome', 'valence']]
```

```python
df.nlargest(10, columns=['duration'])
```

```python
df['contagem_XO'] = df['letra'].str.count('XO')
df_sorted = df.sort_values(by='contagem_XO', ascending=False)
result_df = df_sorted[['nome', 'contagem_XO']]
result_df
```
