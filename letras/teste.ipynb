```python
import pandas as pd
```

```python
df = pd.read_csv("./dadosToSpotify.csv")
df
```

```python
df['contagem_XO'] = df['letra'].str.count('XO')
df_sorted = df.sort_values(by='contagem_XO', ascending=False)
result_df = df_sorted[['nome', 'contagem_XO']]
result_df
```

```python
df.drop_duplicates(subset='nome', keep='first')
df
```

```python
name_counts = df['nome'].value_counts()
repetitive_names = name_counts[name_counts > 1]

print(repetitive_names)
```
